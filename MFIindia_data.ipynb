```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('../siddharth/MFIndia.csv')
```

```python
df.columns = df.columns.str.replace(' ','_')
df
```

```python
print(df.Sector.unique())
sec= input ('Enter Sector name :')
sector= df[df.Sector==sec]
print(sector.category.unique())
cat=input ('Enter category name :')
categ= sector[sector.category==cat]
print(categ.risk_type.unique())
rt= input ('Enter risk_type name :')
risk=categ[categ.risk_type==rt]

fig=plt.figure(figsize=[11,5.5])

a1=fig.add_subplot(121)
sb.lineplot(x=risk.AMC_name,y=risk.return_1yr,label='Return_1yr')
sb.lineplot(x=risk.AMC_name,y=risk.return_3yr,label='Return_3yr')
sb.lineplot(x=risk.AMC_name,y=risk.return_5yr,label='Return_5yr')
plt.xticks(rotation=90)
plt.title('AMC_NAME VS Return,1_yr,3_yr,5_yr')

a2=fig.add_subplot(122)
sb.barplot(x=risk.AMC_name,y=risk.fund_rating,palette='gist_rainbow')
plt.xticks(rotation=90)
plt.title('AMC_NAME VS Fund_Rating')

plt.tight_layout()

plt.show()

```

```python
print(risk.AMC_name.unique())
amc= input ('Enter AMC_name name :')
amcname=risk[risk.AMC_name==amc]

fig=plt.figure(figsize=[17,12])


a3=fig.add_subplot(121)
sb.lineplot(x=amcname.Mutual_Fund_Name,y=amcname.return_1yr,label='Return_1yr')
sb.lineplot(x=amcname.Mutual_Fund_Name,y=amcname.return_3yr,label='Return_3yr')
sb.lineplot(x=amcname.Mutual_Fund_Name,y=amcname.return_5yr,label='Return_5yr')
plt.xticks(rotation=90)
plt.title('Mutual_Fund_Name VS Return,1_yr,3_yr,5_yr')

a4=fig.add_subplot(122)
sb.barplot(x=amcname.Mutual_Fund_Name,y=amcname.fund_rating,palette='gist_rainbow')
plt.xticks(rotation=90)
plt.title('Mutual_Fund_Name VS Fund_Rating')

plt.tight_layout()

plt.show()
```
